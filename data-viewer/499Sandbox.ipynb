```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def read_csv_for_plot(filename):
    averages = []
    parameters = {}
    
    with open(filename, 'r') as inFile:
        firstLine = inFile.readline().strip().split(',')
        secondLine = inFile.readline().strip().split(',')
        
        for i, val in enumerate(firstLine):
            parameters[val] = secondLine[i]
            
        
        trash = inFile.readline()
        
        for i in xrange(int(parameters['numGenerations'])):
            ithLine = inFile.readline()
            
            averages.append(ithLine.split(',')[1])
            
        trash = inFile.readline()
        trash = inFile.readline()
            
    return (averages, parameters)

def plotAllCsvs():
    linewidth = 0.75
    
    csvs = [
            '../runs/saved-runs/01-07-15_d2-00-05/output.csv',
            '../runs/saved-runs/01-07-15_d2-00-10/output.csv',
#             '../runs/saved-runs/01-07-15_d2-00-15/output.csv',
#             '../runs/saved-runs/01-07-15_d2-00-20/output.csv',
#             '../runs/saved-runs/01-07-15_d2-05-05/output.csv',
#             '../runs/saved-runs/01-07-15_d2-05-10/output.csv',
#             '../runs/saved-runs/01-07-15_d2-05-15/output.csv',
#             '../runs/saved-runs/01-07-15_d2-05-20/output.csv',

#             '../runs/saved-runs/01-09-15_d2-00-25/output.csv',
#             '../runs/saved-runs/01-09-15_d2-00-30/output.csv',
#             '../runs/saved-runs/01-09-15_d2-00-35/output.csv',
#             '../runs/saved-runs/01-09-15_d2-00-40/output.csv',
            '../runs/saved-runs/01-09-15_d2-00-45/output.csv',
            '../runs/saved-runs/01-09-15_d2-00-50/output.csv',
#             '../runs/saved-runs/01-09-15_d2-05-25/output.csv',
#             '../runs/saved-runs/01-09-15_d2-05-30/output.csv',

            '../runs/saved-runs/01-03-15_d3-abp-run1/output.csv',
            '../runs/saved-runs/01-03-15_d3-abp-run2/output.csv',
            '../runs/saved-runs/01-03-15_d3-abp-run3/output.csv',
            '../runs/saved-runs/01-03-15_d3-abp-run4/output.csv',
            '../runs/saved-runs/12-29-14_d4-abp/output.csv',
            
            ]
    
    for csv in csvs:
        df = read_csv_for_plot(csv)
    
        plt.plot(df[0], label=df[1]['treeDepth'], lw=linewidth)
        plt.legend(loc=4)
        
plotAllCsvs()
```
